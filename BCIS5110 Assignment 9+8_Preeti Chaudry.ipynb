14. Exame variable 'CURRENCY_EXCHANGE_COVERAGE_RATE'. How many unique values for this variable? How many missing?<br>
Before we rush to replace missing values with mode, we ask the question: is the missing value actually missing or is it something else? <br>
We then read the policies on Kiva about exchange coverage. For instance, a lender gives 25 in US dollars and it changes to 250 in another currency. But when the money is repaid at 250 in the other currency, it only exchanges to 20 US dollars due to the exchange rate fluctuating. There is loss for the lender.<br>
The CURRENCY_EXCHANGE_COVERAGE_RATE measures how much such loss will be covered by the field partner. Some would cover 10% or 20%. So, it is more likely that the missing values indicating the field partner will not share the loss. So, it's more likely missing means 0.<br>

Replace the missing values of this variable with 0.

```python
loans['CURRENCY_EXCHANGE_COVERAGE_RATE'].unique()
```

```python
loans['CURRENCY_EXCHANGE_COVERAGE_RATE'].isna().sum()
```

```python
loans['CURRENCY_EXCHANGE_COVERAGE_RATE']=loans['CURRENCY_EXCHANGE_COVERAGE_RATE'].fillna(0.0)
```

18. Examine the variable 'BORROWER_GENDERS'. There is group borrowing, so we can see multiple borrowers. As research suggests, female borrowers may be riskier than male. It may affect a loan's chance to get funded. We will calculate the percentage of female borrowers and assign it to a new variable. You may write a function and use apply(). 

```python
def perce(value):
    list_all=value.split(',')
    list_all_female=[x for x in list_all if x=='female']
    value_count=len(list_all_female)/len(list_all)
    return value_count
    
```

```python
loans['BORROWER_GENDERS_FEMALE_PERCENTAGE']=loans['BORROWER_GENDERS'].apply(lambda x: perce(str(x)))
```
