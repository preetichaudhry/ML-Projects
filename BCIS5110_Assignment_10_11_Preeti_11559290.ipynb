14. Exame variable 'CURRENCY_EXCHANGE_COVERAGE_RATE'. How many unique values for this variable? How many missing?<br>
Before we rush to replace missing values with mode, we ask the question: is the missing value actually missing or is it something else? <br>
We then read the policies on Kiva about exchange coverage. For instance, a lender gives 25 in US dollars and it changes to 250 in another currency. But when the money is repaid at 250 in the other currency, it only exchanges to 20 US dollars due to the exchange rate fluctuating. There is loss for the lender.<br>
The CURRENCY_EXCHANGE_COVERAGE_RATE measures how much such loss will be covered by the field partner. Some would cover 10% or 20%. So, it is more likely that the missing values indicating the field partner will not share the loss. So, it's more likely missing means 0.<br>

Replace the missing values of this variable with 0.

```python
loans['CURRENCY_EXCHANGE_COVERAGE_RATE'].unique() # only several values
# There are 473884 missing values, about 23% missing.
loans['CURRENCY_EXCHANGE_COVERAGE_RATE'].fillna(0, inplace = True)
```

18. Examine the variable 'BORROWER_GENDERS'. There is group borrowing, so we can see multiple borrowers. As research suggests, female borrowers may be riskier than male. It may affect a loan's chance to get funded. We will calculate the percentage of female borrowers and assign it to a new variable. You may write a function and use apply(). 

```python
def femaleper(genders):
    try:
        genderList = genders.split(',')
        counter = 0
        for gender in genderList:
            if gender == 'female':
                counter += 1
        female = counter/len(genderList)
        return female
    except:
        pass

femaleper('female,female,female,male,male')
```

```python
loans['femalePer'] = loans['BORROWER_GENDERS'].apply(femaleper)
```

28. Prepare y (target) and X (features) using the dataset we get from Q27.

```python
y=all.pop('STATUS')
X=all
```

29. Transform all categorical variables into dummy variables.

```python
X=pd.get_dummies(X)
```

30. Prepare the training and test datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

30. Fit a decision tree classifier.

```python
from sklearn.tree import DecisionTreeClassifier
# Instantiating the model class
regressor = DecisionTreeClassifier()
# Fit the training dataset
regressor.fit(X_train, y_train)
```

31. Get the predicted results for the test dataset

```python
y_prediction = regressor.predict(X_test)
y_prediction
y_prediction.shape
```

32. Evaluate the model using accuracy rate and confusion matrix.

```python
# Model evaluation
# Calculate the accuracy rate
from sklearn.metrics import accuracy_score
accuracy_score(y_true = y_test, y_pred = y_prediction)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true = y_test, y_pred = y_prediction)
```
